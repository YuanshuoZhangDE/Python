```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#数据格式化

datafile = open('Fe-log.lammps', 'r', encoding='utf-8')
datas = datafile.readlines()
#print(type(datas))
datalist = []
for i in range(len(datas)):
    datas[i] = ' '.join(datas[i].split())
    datalist.append(datas[i])

#print(datalist)
    
datafile.close()



datafile = open('Fe-log.txt', 'w')

data = '\n'.join(datalist)

datafile.write(data)

datafile.close()
```

```python
contents = pd.read_csv('Fe-log.txt', sep=' ')

star = int(input("起始行数："))

end = int(input("结束行数："))

contents = contents.iloc[star: end+1, :]

#contents = contents.drop(0)

contents.head()
```

```python
Step = contents['Step']/1000

Temp = contents['Temp']

PE = contents['PotEng']

msd = contents['c_msd_all[4]']

v_twopoint = contents['v_twopoint_all']

#v_fitslope = contents['v_fitslope_all']
```

```python
# Temp
plt.figure()

plt.xlabel("correlation time(ps)")
plt.ylabel("Temp")
plt.plot(Step, Temp)
plt.title('Temp')

plt.show()
```

```python
# Temp
plt.figure()

plt.xlabel("correlation time(ps)")
plt.ylabel("PE")
plt.plot(Step, PE)
plt.title('PE')

plt.show()
```

```python
plt.figure()

plt.xlabel("correlation time(ps)")
plt.ylabel("MSD")
plt.plot(Step, msd, label='msd')
plt.legend(loc="upper left")
plt.title('MSD')

plt.show()
```

```python
plt.figure()

plt.xlabel("correlation time(ps)")
plt.ylabel("Dcoeff")
plt.plot(Step, v_twopoint, linestyle='-', label='end-points_slope')
#plt.plot(Step, v_fitslope_Cr2O3, color="r", linestyle='--', label='fitslope_Cr2O3')
plt.legend(loc="lower left")
plt.title('Dcoeff')

plt.show()
```

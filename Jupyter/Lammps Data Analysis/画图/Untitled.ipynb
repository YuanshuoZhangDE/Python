```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
contents = pd.read_csv('rdf.txt', sep=' ')
contents
```

```python
# RDT 3.5382574
r = contents['r']

gr = contents['gr']

x_smooth = np.linspace(r.min(), r.max(), 50)
y_smooth = make_interp_spline(r, gr)(x_smooth)
```

```python
fig = plt.figure(figsize=(17,5))

# 
fig.add_subplot(1,2,1)
plt.xlabel("r")
plt.ylabel("gr")
plt.plot(r, gr)
plt.title('rdf')

#平滑
fig.add_subplot(1,2,2)
plt.xlabel("r")
plt.ylabel("gr")
plt.plot(x_smooth, y_smooth)
plt.title('rdf')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#数据格式化
datafile = open('log.lammps', 'r', encoding='utf-8')
datas = datafile.readlines()
#print(type(datas))
datalist = []
for i in range(len(datas)):
    datas[i] = ' '.join(datas[i].split())
    datalist.append(datas[i])

#print(datalist)
    
datafile.close()



datafile = open('log.txt', 'w')

data = '\n'.join(datalist)

datafile.write(data)

datafile.close()
```

```python
contents = pd.read_csv('log.txt', sep=' ')

#star = int(input("起始行数："))

#end = int(input("结束行数："))

#contents = contents.iloc[star: end+1, :]

#contents = contents.drop(0)

contents.head()
```

```python
Step = contents['Step']

EBF = contents['EBF']
```

```python
# Temp
plt.figure()

plt.xlabel("neb time(10fs)")
plt.ylabel("EBF")
plt.plot(Step, EBF)
plt.title('neb')

plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1000_cr3 = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/1000k/count_cr3.txt"
t1000_fe = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/1000k/count_fe.txt"
t1000_o = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/1000k/count_o.txt"
t1000_cr = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/1000k/count_cr.txt"
```

```python
t100_cr3_Count = pd.read_csv(t1000_cr3, sep=" ")
t100_fe_Count = pd.read_csv(t1000_fe, sep=" ")
t100_o_Count = pd.read_csv(t1000_o, sep=" ")
t100_cr_Count = pd.read_csv(t1000_cr, sep=" ")
```

```python
position = t100_cr3_Count['Position.Z']

count_cr3 = t100_cr3_Count['Count']
count_fe = t100_fe_Count['Count']
count_o = t100_o_Count['Count']
count_cr = t100_cr_Count['Count']
```

```python
# Count - Position.Z
plt.figure()

plt.xlabel('Position Z')
plt.ylabel('Count')

plt.plot(position, count_cr3, label='cr3')
plt.plot(position, count_o, label='o')
plt.plot(position, count_fe, label='fe')
plt.plot(position, count_cr, label='cr')

plt.legend(loc='upper right')
plt.title('Count Position.Z')

plt.show()
```

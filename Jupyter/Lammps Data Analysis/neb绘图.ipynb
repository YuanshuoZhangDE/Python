```python
import pandas as pd
from pandas.core.frame import DataFrame
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
def neb(ls):
    ls = ls

    print("RD = ",ls[0],"A")

    del ls[0]

    print('RD Energy')

    ls_rd = []
    ls_energy = []

    for i in range(len(ls)):
        if (i % 2) == 0:
            ls_rd.append(ls[i])
        else:
            ls_energy.append(ls[i] - (ls[1]))

    for i in range(len(ls_rd)):
        print(ls_rd[i], ls_energy[i])

    # list格式转DataFrame格式 
    data = {
        'RD' : ls_rd,
        'Energy' : ls_energy
    }

    data = DataFrame(data)

    #绘图
    RD = data['RD']
    Energy = data['Energy']

    x_smooth = np.linspace(RD.min(), RD.max(), 50)
    y_smooth = make_interp_spline(RD, Energy)(x_smooth)

    fig = plt.figure(figsize=(17,5))

    # 
    fig.add_subplot(1,2,1)
    plt.xlabel("distance")
    plt.ylabel("energy")
    plt.plot(RD, Energy)
    plt.title('neb')

#     #平滑
#     fig.add_subplot(1,2,2)
#     plt.xlabel("distance")
#     plt.ylabel("energy")
#     plt.plot(x_smooth, y_smooth)
#     plt.title('neb')
```

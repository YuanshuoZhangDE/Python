```python
t600_stress_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/600k/stress_strain.txt"
t600_msd_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/600k/msd.txt"

t700_stress_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/700k/stress_strain.txt"
t700_msd_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/700k/msd.txt"

t800_stress_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/800k/stress_strain.txt"
t800_msd_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/800k/msd.txt"

t900_stress_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/900k/stress_strain.txt"
t900_msd_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/900k/msd.txt"

t1200_stress_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/1200k/stress_strain.txt"
t1200_msd_path = "C:/Users/Yuanshuo/Desktop/Repositories/Lammps/Cr2O3-FeCr/15w/stress-strain/1200k/msd.txt"
```

```python
t600_Stress = pd.read_csv(t600_stress_path, sep=" ")
t600_Msd = pd.read_csv(t600_msd_path, sep=" ")

t700_Stress = pd.read_csv(t700_stress_path, sep=" ")
t700_Msd = pd.read_csv(t700_msd_path, sep=" ")

t800_Stress = pd.read_csv(t800_stress_path, sep=" ")
t800_Msd = pd.read_csv(t800_msd_path, sep=" ")

t900_Stress = pd.read_csv(t900_stress_path, sep=" ")
t900_Msd = pd.read_csv(t900_msd_path, sep=" ")

t1200_Stress = pd.read_csv(t1200_stress_path, sep=" ")
t1200_Msd = pd.read_csv(t1200_msd_path, sep=" ")
```

```python
step = np.arange(1001)
strain = t600_Stress['strain']
t600_stress = t600_Stress['stress']
t700_stress = t700_Stress['stress']
t800_stress = t800_Stress['stress']
t900_stress = t900_Stress['stress']
t1200_stress = t1200_Stress['stress']

t600_msd_cr3 = t600_Msd['cr3']
t600_msd_o = t600_Msd['o']
t600_msd_fe = t600_Msd['fe']
t600_msd_cr = t600_Msd['cr']

t700_msd_cr3 = t700_Msd['cr3']
t700_msd_o = t700_Msd['o']
t700_msd_fe = t700_Msd['fe']
t700_msd_cr = t700_Msd['cr']

t800_msd_cr3 = t800_Msd['cr3']
t800_msd_o = t800_Msd['o']
t800_msd_fe = t800_Msd['fe']
t800_msd_cr = t800_Msd['cr']

t900_msd_cr3 = t900_Msd['cr3']
t900_msd_o = t900_Msd['o']
t900_msd_fe = t900_Msd['fe']
t900_msd_cr = t900_Msd['cr']

t1200_msd_cr3 = t1200_Msd['cr3']
t1200_msd_o = t1200_Msd['o']
t1200_msd_fe = t1200_Msd['fe']
t1200_msd_cr = t1200_Msd['cr']
```

```python
# stress-strain
plt.figure()

plt.xlabel('Strain')
plt.ylabel('Stress(Gpa)')

plt.plot(strain, t600_stress, label='600k')
plt.plot(strain, t700_stress, label='700k')
plt.plot(strain, t800_stress, label='800k')
plt.plot(strain, t900_stress, label='900k')
plt.plot(strain, t1200_stress, label='1200k')

plt.legend(loc='upper left')
plt.title('Stress-Strain')

plt.show()
```

```python
# 600k msd
plt.figure()

plt.xlabel('Time(0.1ps)')
plt.ylabel('MSD')

plt.plot(step, t600_msd_cr3, label='cr3')
plt.plot(step, t600_msd_o, label='o')
plt.plot(step, t600_msd_fe, label='fe')
plt.plot(step, t600_msd_cr, label='cr')

plt.legend(loc='upper left')
plt.title('MSD 600k')

plt.show()
```

```python
# 700k msd
plt.figure()

plt.xlabel('Time(0.1ps)')
plt.ylabel('MSD')

plt.plot(step, t700_msd_cr3, label='cr3')
plt.plot(step, t700_msd_o, label='o')
plt.plot(step, t700_msd_fe, label='fe')
plt.plot(step, t700_msd_cr, label='cr')

plt.legend(loc='upper left')
plt.title('MSD 700k')

plt.show()
```

```python
# Cr msd
plt.figure()

plt.xlabel('Time(0.1ps)')
plt.ylabel('MSD')

plt.plot(step, t600_msd_cr, label='600k')
plt.plot(step, t700_msd_cr, label='700k')
plt.plot(step, t800_msd_cr, label='800k')
plt.plot(step, t900_msd_cr, label='900k')
plt.plot(step, t1200_msd_cr, label='1200k')

plt.legend(loc='upper left')
plt.title('Cr MSD')

plt.show()
```

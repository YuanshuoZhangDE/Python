```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
contents = pd.read_csv('neb.txt', sep=' ')
contents
```

```python
# RDT 3.5382574
RD = contents['RD']

Energy = contents['Energy']

x_smooth = np.linspace(RD.min(), RD.max(), 50)
y_smooth = make_interp_spline(RD, Energy)(x_smooth)
```

```python
fig = plt.figure(figsize=(17,5))

# 
fig.add_subplot(1,2,1)
plt.xlabel("distance")
plt.ylabel("energy")
plt.plot(RD, Energy)
plt.title('neb')

#平滑
fig.add_subplot(1,2,2)
plt.xlabel("distance")
plt.ylabel("energy")
plt.plot(x_smooth, y_smooth)
plt.title('neb')
```

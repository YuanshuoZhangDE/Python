```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from matplotlib import rc
```

```python
contents = pd.read_csv('C:/Users/Yuanshuo/Desktop/应力应变.txt', sep=' ')
contents
```

```python
rc('mathtext', default='regular')

strain = contents['strain']
stress = contents['stress']
dislocation = contents['dislocation']

fig = plt.figure()
ax = fig.add_subplot(111)
lns1 = ax.plot(strain, stress, color='r', label='stress')

ax2 = ax.twinx()
lns2 = ax2.plot(strain, dislocation, label='dislocation')

lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc='lower right')


ax.grid()
ax.set_xlabel('Strain')
ax.set_ylabel('Stress (Gpa)')
ax2.set_ylabel('Total dislocation length (A)')
```
